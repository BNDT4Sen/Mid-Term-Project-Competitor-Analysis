```python
# Define API key and endpoint
url = 'http://api.citybik.es/v2/networks'

# Define the parameters for request
params = {
    'query': 'restaurants in Vancouver',
    'radius': 1000,  # Radius in meters
    'key': city_bike
}

# Send the initial request
response = requests.get(url, params=params)

# Parse the response as JSON
data = response.json()
data
```

```python
url = 'http://api.citybik.es/v2/networks/mobibikes'
response = requests.get(url)
print(response)
```

```python
# parse through response to get the details
response.json()['network']
```

```python
station_list = response.json()['network']['stations']
station_list[0]
```

```python
name = [station['name'] for station in station_list]
latitude = [station['latitude']for station in station_list]
longitude = [station['longitude'] for station in station_list]
bike_slots = [station['extra']['slots'] for station in station_list]
ebike = [station['extra']['ebikes']for station in station_list]
```

```python
city_bike =  pd.DataFrame(data={'station_name': name, 'latitude': latitude, 'longitude': longitude, 'bikes': bike_slots, 'ebike':ebike})
```

```python
city_bike['station_location'] = city_bike.apply(lambda row: f"{row['latitude']:.6f}, {row['longitude']:.6f}", axis=1)
city_bike
```
