```python
import os
import requests
import haversine as hs   
from haversine import haversine, Unit
google_place = os.environ["GOOGLE_API"]
```

```python

# Function to calculate the distance between two sets of latitude and longitude coordinates using the haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    return haversine((lat1, lon1), (lat2, lon2), unit=Unit.KILOMETERS)

# Calculate the distance between each restaurant and SkyTrain station
distances = []
for index, row in merged_venues.iterrows():
    restaurant_lat = row['Latitude_x']
    restaurant_lng = row['Longitude_x']
    station_lat = row['Latitude_y']
    station_lng = row['Longitude_y']
    
    distance = calculate_distance(restaurant_lat, restaurant_lng, station_lat, station_lng)
    distances.append(distance)

# Add the distance column to the DataFrame
merged_venues['Distance to Station'] = distances


merged_venues
```
