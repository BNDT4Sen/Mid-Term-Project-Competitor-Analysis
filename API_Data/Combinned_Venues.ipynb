```python
import os
import requests
import haversine as hs   
from haversine import haversine, Unit
google_place = os.environ["GOOGLE_API"]
```

```python
# Define your API key and endpoint
url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

# Define the location (Vancouver coordinates)
location = '49.2827,-123.1207'

# Define the radius in meters (1000 meters)
radius = 1000

# Define the type as 'restaurant'
place_type = 'restaurant'

# Define the parameters for your request
params = {
    'location': location,
    'radius': radius,
    'type': place_type,
    'key': google_place
}

# List to store the restaurant and SkyTrain station data
data_list = []

# Send the request to get restaurants
response = requests.get(url, params=params)

# Parse the response as JSON
data = response.json()

# Handle the response data
if response.status_code == 200:
    restaurants = data['results']

    # Iterate over the restaurants
    for restaurant in restaurants:
        restaurant_name = restaurant['name']
        restaurant_location = restaurant['geometry']['location']
        restaurant_rating = restaurant.get('rating', 'N/A')
        restaurant_lat = restaurant_location['lat']
        restaurant_lng = restaurant_location['lng']
        restaurant_address = restaurant['vicinity']
        restaurant_id = restaurant['place_id']

        restaurant_data = {
            'Type': 'Restaurant',
            'Name': restaurant_name,
            'Rating': restaurant_rating,
            'Latitude': restaurant_lat,
            'Longitude': restaurant_lng,
            'Address': restaurant_address,
            'ID': restaurant_id
        }

        # Send a new request to find nearby SkyTrain stations for each restaurant
        skytrain_params = {
            'location': f"{restaurant_lat},{restaurant_lng}",
            'radius': radius,
            'keyword': 'skytrain station',
            'key': google_place
        }

        skytrain_response = requests.get(url, params=skytrain_params)
        skytrain_data = skytrain_response.json()

        if skytrain_response.status_code == 200:
            skytrain_results = skytrain_data['results']
            for skytrain_station in skytrain_results:
                station_name = skytrain_station['name']
                station_location = skytrain_station['geometry']['location']
                station_lat = station_location['lat']
                station_lng = station_location['lng']
                station_address = skytrain_station['vicinity']
                station_id = skytrain_station['place_id']

                station_data = {
                    'Type': 'SkyTrain Station',
                    'Name': station_name,
                    'Rating': 'N/A',
                    'Latitude': station_lat,
                    'Longitude': station_lng,
                    'Address': station_address,
                    'ID': station_id
                }

                # Calculate the distance between the restaurant and SkyTrain station
                restaurant_coord = (restaurant_lat, restaurant_lng)
                station_coord = (station_lat, station_lng)
                distance = haversine(restaurant_coord, station_coord, unit=Unit.KILOMETERS)
                station_data['Distance'] = distance

                # Append the restaurant and station data to the list
                data_list.append(restaurant_data)
                data_list.append(station_data)

# Create a DataFrame from the data list
df = pd.DataFrame(data_list)
```
