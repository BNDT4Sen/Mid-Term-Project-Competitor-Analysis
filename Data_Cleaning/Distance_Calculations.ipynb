```python
#Package called Haversine is imported to simplify some otherwise complicated sphereical geometry
import pandas as pd
import haversine as hs
from haversine import Unit
```

```python
#DataFrames previously created with data via API requests are called

SkyTrain_DF = pd.read_csv('skytrain_stations.csv')

Google_DF = pd.read_csv('google_venues.csv')

Yelp_DF = pd.read_csv('yelp_venues.csv')
```

```python
#Each venue is iterated on each station to calculate and record the distance between the two.
distance_list = []
for station_index, station in SkyTrain_DF.iterrows():
    for venue_index, venue in Google_DF.iterrows():
        if venue['SkyTrain Station'] == station['Name']:
            #Grabbing Latitudes and Longitudes for venue and station.
            station_Latitude = station['Latitude']
            station_Longitude = station['Longitude']
            venue_Latitude = venue['Latitude']
            venue_Longitude = venue['Longitude']
            #Combining coordinates into a format that haversine can use.
            station_Location = (station_Latitude, station_Longitude)
            venue_Location = (venue_Latitude, venue_Longitude)
            #Using haversine to calculate distances between the coordinates, and recording to distance_list.
            distance_from_station = hs.haversine(station_Location, venue_Location, unit = Unit.METERS)
            distance_list.append(distance_from_station)
# Distance data is appended to the DataFrame. 
Google_DF['Distance'] = distance_list
```

```python
#The Series in the DataFrame are sorted based on distance. Duplicate venues with greater distances are then removed.  
google_venues_sorted = Google_DF.sort_values(by = ['Distance'])
google_venues_cleaning = google_venues_sorted.drop_duplicates(subset = ['ID', 'Venue Name', 'Address'], keep = 'first', ignore_index = True)
#Standardizing the DataFrame prior to merging.
google_pre_merge = google_venues_cleaning.drop(labels = ['Address'], axis = 1)
#DataFrame is saved as csv and retrieved in merging_venues_data.ipynb
google_pre_merge.to_csv('google_pre_merge.csv', index=False)
```
